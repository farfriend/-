```python
import gym
import time

env = gym.make('MountainCar-v0')

for episode in range(3):
    observation = env.reset()
    total_reward = 0
    print(f"\n=== ç¬¬ {episode+1} å›åˆ ===")
    
    for step in range(500):
        # æ¸²æŸ“ç¯å¢ƒï¼ˆè¿™ä¼šå¼¹å‡ºçª—å£æ˜¾ç¤ºï¼‰
        env.render()
        
        # æ§åˆ¶æ¸²æŸ“é€Ÿåº¦ï¼ˆå•ä½ï¼šç§’ï¼‰
        time.sleep(0.02)  # 0.02ç§’ = 50FPS
        
        # æ‰§è¡ŒåŠ¨ä½œ
        action = env.action_space.sample()  # éšæœºåŠ¨ä½œ
        # æˆ–è€…ä½¿ç”¨ç®€å•ç­–ç•¥ï¼š
        # position, velocity = observation
        # action = 0 if velocity < 0 else 2  # ç®€å•ç­–ç•¥
        
        observation, reward, terminated, truncated, info = env.step(action)
        total_reward += reward
        
        # æ˜¾ç¤ºå½“å‰çŠ¶æ€ä¿¡æ¯
        position, velocity = observation
        print(f"æ­¥éª¤ {step}: ä½ç½®={position:.3f}, é€Ÿåº¦={velocity:.3f}, åŠ¨ä½œ={action}", end='\r')
        
        if terminated:
            print(f"\nğŸ‰ æˆåŠŸï¼åœ¨ç¬¬ {step+1} æ­¥åˆ°è¾¾ç›®æ ‡ï¼")
            break
        elif truncated:
            print(f"\nâ° åœ¨ç¬¬ {step+1} æ­¥ç»“æŸï¼Œç´¯è®¡å¥–åŠ±: {total_reward}")
            break

env.close()
```
